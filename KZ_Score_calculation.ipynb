```python
import subprocess

# 参数
batch_size = 100  # 每个批次处理的文件数
chromosome = 19  # 要分析的染色体号
bgen_file_id = "file-FxY60F8JkF63k38X9V5BFY55"  # bgen 文件 ID
bgi_file_id = "file-FxZ2byjJkF65g2vX9Vx8vgb3"  # bgi 文件 ID
sample_file_id = "file-GbxY6YQJ5YqBqx9qX6Yyv4Bk"  # sample 文件 ID
destination_folder = "project-GbjJ860J2p5ZJqZKZZVY883Z:/omics_pred/score_files1201/"  # 目标目录
file_mapping_path = "file_name_id_mapping.txt"  # 文件名与 ID 映射文件路径

# 读取文件名和文件 ID 的映射
file_mapping = []
with open(file_mapping_path, "r") as f:
    lines = f.readlines()

for line in lines:
    file_id, file_name = line.strip().split("\t")  # 确保文件名在第二列
    file_mapping.append((file_name, file_id))

# 筛选出指定染色体的文件
score_files_chr = []
for file_name, file_id in file_mapping:
    if f"_{chromosome}." in file_name:  # 根据文件名中的染色体号筛选
        score_files_chr.append((file_name, file_id))

if not score_files_chr:
    print("No files matched the specified chromosome.")
    exit(1)

# 验证文件格式并过滤无效文件
valid_score_files = []
skipped_files = []  # 保存跳过的文件
for file_name, file_id in score_files_chr:
    try:
        # 使用 dx cat 获取文件内容
        result = subprocess.run(
            ["dx", "cat", f"project-GbjJ860J2p5ZJqZKZZVY883Z:{file_id}"],
            capture_output=True,
            text=True,
            check=True
        )
        lines = result.stdout.splitlines()
        if len(lines) < 2:  # 文件内容少于两行（空文件或只有一个 rsid）
            print(f"Skipping invalid score file (too few lines): {file_name}")
            skipped_files.append(file_name)
        else:
            valid_score_files.append((file_name, file_id))
    except Exception as e:
        print(f"Skipping invalid score file (error reading file): {file_name}, error: {e}")
        skipped_files.append(file_name)

if not valid_score_files:
    print("No valid score files found.")
    exit(1)

# 分批处理文件
batches = [
    valid_score_files[i:i + batch_size]
    for i in range(0, len(valid_score_files), batch_size)
]

# 对每个批次生成和执行命令
for batch_idx, batch in enumerate(batches):
    print(f"Processing batch {batch_idx + 1}/{len(batches)}")

    # 构造 -iin 参数和 plink2 命令
    iin_params = []
    plink_commands = []

    for file_name, file_id in batch:
        iin_params.append(f"-iin=project-GbjJ860J2p5ZJqZKZZVY883Z:{file_id}")
        plink_commands.append(
            f"plink2 --bgen ukb22828_c{chromosome}_b0_v3.bgen ref-first "
            f"--sample ukb22828_c{chromosome}_b0_v3.sample "
            f"--score {file_name} 1 4 6 list-variants cols=scoresums "
            f"--out {file_name.split('.')[0]} || true"
        )

    # 合并参数
    iin_params_str = " ".join(iin_params)
    plink_commands_str = " && ".join(plink_commands)

    # 跳过空命令的情况
    if not plink_commands_str.strip():
        print(f"Batch {batch_idx + 1}: No valid plink commands generated.")
        continue

    # 构造 dx 命令
    dx_command = f"""
    dx run swiss-army-knife \\
        --priority high \\
        --instance-type "mem3_ssd1_v2_x16" \\
        -iin=project-GbjJ860J2p5ZJqZKZZVY883Z:{bgen_file_id} \\
        -iin=project-GbjJ860J2p5ZJqZKZZVY883Z:{bgi_file_id} \\
        -iin=project-GbjJ860J2p5ZJqZKZZVY883Z:{sample_file_id} \\
        {iin_params_str} \\
        -icmd="{plink_commands_str}" \\
        --destination "{destination_folder}" \\
        --tag plink2_score_batch{batch_idx + 1} \\
        -y
    """
    
    # 打印并运行命令
    print(f"Executing batch {batch_idx + 1}/{len(batches)}:")
    print(dx_command)
    try:
        subprocess.run(dx_command, shell=True, check=True)
    except subprocess.CalledProcessError as e:
        print(f"Error in batch {batch_idx + 1}: {e}")

# 打印跳过的文件列表
if skipped_files:
    print(f"Skipped {len(skipped_files)} invalid files:")
    for skipped_file in skipped_files:
        print(f"  - {skipped_file}")
```
